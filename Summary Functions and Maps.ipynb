```python
# Quick Overview

# describe()
# mean()
# unique()
# value_counts() - unique values and how often they occur

# Maps
# A fucntion that takes one set of values and maps them to another set of values.
#2 methods
# map()
## returns a Series

# apply()
## returns a dataframe
## equivalent method to transform a whole DF by calling a custom method on each row.
## axis='columns' means transformation is applied to each row.
## axis='index' means transformation is applied to each column.

#map() and apply() do not modify the original data. They return transformed Series and DataFrames respectively.

# pandas provides common mapping operations as built-ins. These are faster as they use speed ups built into pandas.
```

```python
import pandas as pd

#read housing dataset
X = pd.read_csv("train.csv", index_col='Id')

X.head()
```

```python
#remean SalePrice to 0 using map()

SalePrice_mean = X.SalePrice.mean()
SalePrice_mean
X.SalePrice.map(lambda p: p - SalePrice_mean)
```

```python
# using the apply() method to remean SalePrice to 0

def remean_price(row):
    row.SalePrice = row.SalePrice - SalePrice_mean
    return row
X.apply(remean_price, axis='columns')
```

```python
# Better way to calculate mean without using map() and apply()
# pandas understands we are subtracting a constant value from a series of prices

SalePrice_mean = X.SalePrice.mean()
X.SalePrice - SalePrice_mean
```

```python
# Divide SalePrice into 3 categories - 'inexpensive', 'fair value', 'expensive'

#X.SalePrice.describe()
SalePrice_25qtle = X.SalePrice.quantile(0.25)
SalePrice_75qtle = X.SalePrice.quantile(0.75)

def price_to_cat(row):
    if row.SalePrice <= SalePrice_25qtle:
        return 'inexpensive'
    elif row.SalePrice >= SalePrice_75qtle:
        return 'expensive'
    else:
        return 'fair value'
    
price_cat = X.apply(price_to_cat, axis='columns')
price_cat
```

```python
#X_train = X.assign(price_cat = price_cat)
# axis number -> 0 = rows, 1 = columns
#X_train.drop('price_cat', axis=1, inplace=True)
```

```python
# create a new column with the ratio of prict to area.
# Use this ratio to divide houses into 3 categories - inexpensive, fair value and expensive ( in the next cell)
X_train = X.assign(price_to_area = X.SalePrice/X.LotArea)
X_train
```

```python
X_train.price_to_area.idxmin()
```

```python
# Divide price_to_area into 3 categories - 'inexpensive', 'fair value', 'expensive'

#X.SalePrice.describe()
p2a_25qtle = X_train.price_to_area.quantile(0.25)
p2a_75qtle = X_train.price_to_area.quantile(0.75)

def p2a_to_cat(row):
    if row.price_to_area <= p2a_25qtle:
        return 'bargain'
    elif row.price_to_area >= p2a_75qtle:
        return 'money_printer_go_brr'
    else:
        return 'fair_value'
    
bargain_cat = X_train.apply(p2a_to_cat, axis='columns')
bargain_cat.value_counts()
```

```python
X_train.price_to_area.describe()
```

```python
X_train = X_train.assign(bargain_cat = bargain_cat)
X_train
```
